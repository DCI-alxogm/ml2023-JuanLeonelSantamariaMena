```python
def f(x,w):
  y=w[0]
  for i in range(1,len(w)):
    y += w[i]*x[i-1]
  return y

def cost_(x,y,w,f):
  cost= np.sum((f(x,w)-y)**2)/(2*len(x[0]))
  return cost

def grad(x,y,w,f):
  dw=np.zeros(len(w))
  fw= f(x,w)
  dw[0]=np.sum((fw-y))/len(y)
  for j in range(1,len(w)):
    dw[j]=np.sum((fw-y)*x[j-1])/len(y)
  return dw

def grad_des(x,y,w,alpha,iter,cost_,grad,f):
  cost=[]
  for i in range(iter):
    dj = grad(x,y,w,f)
    for j in range(len(w)):
      w[j]=w[j]-alpha*dj[j]
    cost.append(cost_(x,y,w,f))
    if i % math.ceil(iter / 10) == 0:
            print(f"Iteración {i:4}: w[0]: {w[0]:0.5e}, w[1]: {w[1]:0.5e}, w[2]:{w[2]:0.5e},w[3]:{w[3]:0.5e},w[4]:{w[4]:0.5e}")
  return w , cost
```
