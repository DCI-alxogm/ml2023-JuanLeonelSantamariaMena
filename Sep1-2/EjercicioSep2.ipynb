```python
def f(x,w):
  y=w[0]
  for i in range(1,len(w)):
    y += w[i]*x[i-1]
  return y

def cost_(x,y,w):
  cost= f(x,w)/(2*len(x))
  return cost

def grad(x,y,w):
  dw=np.zeros(len(w))
  fw=w[0]
  for i in range(1,len(w)):
    fw+=w[i]*x[i-1]
  dw[0]=np.sum((fw-y))/len(y)
  for j in range(1,len(w)):
    dw[j]=np.sum((fw-y)*x[j-1])/len(y)
  return dw

def wi(dw,w,alpha):
  wi=np.zeros(len(w))
  for i in range(len(w)):
    wi[i]=w[i]-alpha*dw[i]
  return wi

def grad_des(x,y,w,alpha,iteracines):
  for i in range(iteraciones):
    dw = grad(x,y,w)
    w = wi(dw,w,alpha)
  return w
```

```python
#pesos
w0=0.0
w1=0.0
w2=0.0
w3=0.0
w4=0.0
w=np.array([w0,w1,w2,w3,w4])
#
iteraciones=10000
alpha=1*10^(-5)

grad_des(Xh,th,w,alpha,iteraciones)
```
